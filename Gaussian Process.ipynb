```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
s=100
x = np.linspace(0,1,s)[:,None]
m = np.zeros(s)
x1 = np.tile(x,s)
x2 = x1.T
S = np.exp(-np.power((x1-x2),2))
S2 = (1+ np.abs(x1-x2))*np.exp(-np.abs(x1-x2))
for i in range(5):
    y = np.random.multivariate_normal(mean=m, cov=S2)
    plt.plot(x,y)
```
