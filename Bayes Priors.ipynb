```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 0
s = 0.4
x = np.linspace(-2,4,100)
plt.figure(figsize=(12,8))
for i in range(100):
    w1 = np.random.normal(m, s)
    w0 = np.random.normal(m, 3*s)
    y = w1*x + w0
    plt.plot(x,y, color='grey')
plt.ylim(-5,5)
```

```python
p=np.zeros(10000)
for i in range(10000):
    m = np.random.normal(0, 1)
    ms = np.random.normal(m, 1, 12)
    p[i] = (np.abs(ms.mean()) -  np.sqrt(ms.var()))
```
